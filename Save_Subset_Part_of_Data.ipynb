### Importing Libraries

```python
import pandas as pd
import numpy as np
```

### Read Titanic Dataset From Link

```python
df = pd.read_csv("https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv", index_col="PassengerId")
```

```python
df.shape
```

```python
df.head()
```

### Getting Subset of The Data

#### Save 3 Columns & Last 3 Records

```python
subset_data = df.tail(3)[["Name", "Pclass", "Embarked"]]
```

```python
subset_data.to_csv("last_3_records_output.csv")
```

### Getting Subset According To Condition

```python
# Get All Male That Age > 35 and Survived
male_survived_filter = (df["Sex"] == "male") & (df["Age"] > 35) & (df["Survived"] == 1)
male_survived = df.loc[male_survived_filter, ["Name", "Age", "Embarked"]]
male_survived
```

```python
male_survived.to_csv("survived_male_output.csv", index=False)
```
